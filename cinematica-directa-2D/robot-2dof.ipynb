# Cinemática directa para robot de 2 grados de libertad

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def fwd_kinematics(q1, q2):
    '''
    Implements the forward kinematics of the robot. 
    
    Arguments
    ---------
    q1 : float
       Angle in radians or position in meter of the first degree of freedom
    q2 : float
       Angle in radians or position in meter of the second degree of freedom
       
    Returns
    -------
    x : float
       The position in the global x-direction of the end-effector (tool point)
    y : float
       The position in the global y-direction of the end-effector (tool point)
    theta : float
       The orientation of the end-effector with respect to the positive global x-axis
    '''
    x = np.random.randn(1)
    y = np.random.randn(1)
    theta = np.random.randn(1)
    return (x, y, theta)
```

## Genera puntos en el espacio de las articulaciones y visualiza el espacio de trabajo del robot

```python
q1 = np.arange(0, 2*np.pi, 0.1)
q2 = np.arange(-np.pi, np.pi, 0.1)
N = len(q1)
pose = np.empty((N,N,3))
for i1_ in range(N):
    for i2_ in range(N):
        pose[i1_, i2_] = fwd_kinematics(q1_, q2_)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(pose[:,:,0], pose[:,:,1], pose[:,:,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('theta')
```
